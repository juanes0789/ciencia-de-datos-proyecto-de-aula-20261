1. Carga el conjunto de datos **wine** desde la librería `sklearn.datasets` usando `load_wine()`.

2. Convierte el conjunto de datos en un **DataFrame de pandas** e identifica las variables disponibles.

3. Selecciona al menos **tres variables numéricas** para el análisis (por ejemplo):
   - `alcohol`
   - `malic_acid`
   - `color_intensity`

4. Aplica los siguientes métodos de escalamiento sobre las variables seleccionadas:
   - `StandardScaler`
   - `MinMaxScaler`
   - `RobustScaler`
   - `Z-Score`

5. Genera **gráficas comparativas** que permitan visualizar la distribución de al menos **una de las variables seleccionadas** en los siguientes escenarios:
   - Distribución **original**
   - Distribución con **StandardScaler**
   - Distribución con **MinMaxScaler**

6. Escribe un análisis de **máximo cinco líneas** en el que expliques las diferencias entre las técnicas o métodos, su impacto en los resultados y cuál consideras más recomendable, justificando brevemente tu elección.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine

datos = load_wine()
df_wine = pd.DataFrame(datos.data, columns=datos.feature_names)
df_wine.head()
```

```python
df_wine.info()
```

Variables escogidas:
- alcohol
- color_intensity
- proline

## Escalamiento

```python
# StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_standard = scaler.fit_transform(df_wine[['alcohol', 'color_intensity', 'proline']])
df_standard = pd.DataFrame(df_standard, columns=['alcohol', 'color_intensity', 'proline'])
df_standard.head()
```

```python
# MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_minMax = scaler.fit_transform(df_wine[['alcohol', 'color_intensity', 'proline']])
df_minMax = pd.DataFrame(df_minMax, columns=['alcohol', 'color_intensity', 'proline'])
df_minMax.head()
```

```python
# RobustScaler
from sklearn.preprocessing import RobustScaler
rScaler = RobustScaler()

df_rs = rScaler.fit_transform(df_wine[['alcohol', 'color_intensity', 'proline']])
df_rs = pd.DataFrame(df_rs, columns=['alcohol', 'color_intensity', 'proline'])
df_rs.head()
```

```python
# Z-Score

from scipy.stats import zscore
# La función zscore de scipy.stats se usa directamente.
scaled_data = zscore(df_wine[['alcohol', 'color_intensity', 'proline']])
df_zscore = pd.DataFrame(scaled_data, columns=['alcohol', 'color_intensity', 'proline'])
df_zscore.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(18, 6))

# Original Distribution
plt.subplot(1, 5, 1)
sns.histplot(df_wine['alcohol'], kde=True)
plt.title('Original Alcohol')

# StandardScaler Distribution
plt.subplot(1, 5, 2)
sns.histplot(df_standard['alcohol'], kde=True)
plt.title('StandardScaler Alcohol')

# MinMaxScaler Distribution
plt.subplot(1, 5, 3)
sns.histplot(df_minMax['alcohol'], kde=True)
plt.title('MinMaxScaler Alcohol')

# RobustScaler Distribution
plt.subplot(1, 5, 4)
sns.histplot(df_rs['alcohol'], kde=True)
plt.title('RobustScaler Alcohol')

# Z-Score Distribution
plt.subplot(1, 5, 5)
sns.histplot(df_zscore['alcohol'], kde=True)
plt.title('Z-Score Alcohol')

plt.tight_layout()
plt.show()
```

El **StandardScaler** y **Z-Score** producen distribuciones similares, centradas en 0 con desviación estándar 1, siendo sensibles a outliers. El **MinMaxScaler** escala los datos a un rango [0, 1], manteniendo la forma original. El **RobustScaler** es más resistente a outliers, ya que usa la mediana e IQR. Para datos con muchos outliers, RobustScaler es recomendable; de lo contrario, StandardScaler suele ser una buena elección. Las distribuciones visualizadas confirman estos efectos, con StandardScaler y Z-Score presentando una escala similar para 'alcohol', mientras que MinMaxScaler comprime los valores a un rango de 0 a 1 y RobustScaler los ajusta alrededor de la mediana, demostrando la resistencia a valores extremos.

2. Convierte el conjunto de datos en un **DataFrame de pandas** e identifica las variables disponibles.

4. Aplica los siguientes métodos de **normalización** sobre las variables seleccionadas utilizando `Normalizer` de `sklearn.preprocessing`:
   - **Normalización L1**
   - **Normalización L2**

5. Genera **gráficas comparativas** que permitan visualizar la distribución de al menos **una de las variables seleccionadas** en los siguientes escenarios:
   - Distribución **original**
   - Distribución con **normalización L1**
   - Distribución con **normalización L2**

6. Escribe un análisis de **máximo cinco líneas** en el que expliques las diferencias entre las técnicas o métodos, su impacto en los resultados y cuál consideras más recomendable, justificando brevemente tu elección.

```python
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import Normalizer

housing = fetch_california_housing()
df_housing = pd.DataFrame(housing.data, columns=housing.feature_names)
df_housing.head()
```

```python
df_housing.info()
```

## Variables escogidas

- MedInc
- AveRooms
- HouseAge

```python
vars_housing = ['MedInc', 'AveRooms', 'HouseAge']

normalizer_l1 = Normalizer(norm='l1')
normalizer_l2 = Normalizer(norm='l2')

df_l1 = pd.DataFrame(
    normalizer_l1.fit_transform(df_housing[vars_housing]),
    columns=vars_housing
)
df_l2 = pd.DataFrame(
    normalizer_l2.fit_transform(df_housing[vars_housing]),
    columns=vars_housing
)

df_l1.head()
```

## Gráficos comparativos para MedInc

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.histplot(df_housing['MedInc'], kde=True)
plt.title('Distribución original de MedInc')

plt.subplot(1, 3, 2)
sns.histplot(df_l1['MedInc'], kde=True)
plt.title('MedInc con normalización L1')

plt.subplot(1, 3, 3)
sns.histplot(df_l2['MedInc'], kde=True)
plt.title('MedInc con normalización L2')

plt.tight_layout()
plt.show()
```

La normalización **L1** divide cada fila por la suma de valores absolutos y genera distribuciones más comprimidas, preservando la proporcionalidad relativa. La normalización **L2** divide por la norma euclidiana, penalizando más los valores grandes y suavizando su influencia. Ambas operan por fila, por lo que cambian la escala en función del vector completo de cada observación. Los gráficos de 'MedInc' muestran cómo ambas transformaciones comprimen la distribución de la variable, siendo la L1 la más radical en este aspecto. Para modelos basados en distancia, L2 suele ser preferible; L1 es útil cuando interesa robustez relativa frente a magnitudes dispersas.

## <span style="color:#2F749F;"><strong>Ejercicio 4: Comparación de transformaciones para reducir asimetría</strong></span>

A partir de un conjunto de datos real sobre **pacientes con diabetes**, analiza cómo cambian las distribuciones de algunas variables clínicas al aplicar diferentes **transformaciones matemáticas utilizadas para reducir la asimetría (skewness) o estabilizar la varianza**.

1. Carga el conjunto de datos **diabetes** desde la librería `sklearn.datasets` usando `load_diabetes()`.

2. Convierte el conjunto de datos en un **DataFrame de pandas** e identifica las variables disponibles.

3. Selecciona al menos **dos variables numéricas** para el análisis (por ejemplo: `bmi`, `bp` o `s5`).

4. Aplica las siguientes **transformaciones** sobre una de las variables seleccionadas:

   - **Transformación logarítmica** usando `np.log()` o `np.log1p()`
   - **Transformación logarítmica reflejada** (útil cuando existe asimetría negativa)
   - **Transformación cuadrática** usando `x²`
   - **Transformación Box-Cox** usando `scipy.stats.boxcox`

5. Genera **gráficas comparativas** que permitan visualizar la distribución de la variable en los siguientes escenarios:
   - Distribución **original**
   - Distribución con **transformación log**
   - Distribución con **log reflejado**
   - Distribución con **transformación cuadrática**
   - Distribución con **transformación Box-Cox**

6. Escribe un análisis de **máximo cinco líneas** en el que expliques las diferencias entre las técnicas o métodos, su impacto en los resultados y cuál consideras más recomendable, justificando brevemente tu elección.

```python
from scipy.stats import boxcox
from sklearn.datasets import load_diabetes

datos_diabetes = load_diabetes()
df_diabetes = pd.DataFrame(datos_diabetes.data, columns=datos_diabetes.feature_names)
df_diabetes.head()
```

```python
df_diabetes.info()
```

## Variables escogidas

- bmi
- bp
- s5

```python
# Se transforma la variable 's5' para comparar efectos sobre la asimetría.
x_original = df_diabetes['s5'].copy()

# Ajuste a positivo para aplicar log y Box-Cox de forma segura
x_pos = x_original - x_original.min() + 1e-6

x_log = np.log(x_pos)
x_reflejada_log = np.log((x_original.max() - x_original) + 1e-6)
x_cuadratica = x_original ** 2
x_boxcox, lambda_boxcox = boxcox(x_pos)

df_transf_diabetes = pd.DataFrame({
    'original': x_original,
    'log': x_log,
    'log_reflejada': x_reflejada_log,
    'cuadratica': x_cuadratica,
    'boxcox': x_boxcox
})

df_transf_diabetes.head()
```

## Gráficos comparativos para s5

```python
plt.figure(figsize=(20, 5))

plt.subplot(1, 5, 1)
sns.histplot(df_transf_diabetes['original'], kde=True)
plt.title('Original s5')

plt.subplot(1, 5, 2)
sns.histplot(df_transf_diabetes['log'], kde=True)
plt.title('Log(s5 ajustada)')

plt.subplot(1, 5, 3)
sns.histplot(df_transf_diabetes['log_reflejada'], kde=True)
plt.title('Log reflejada')

plt.subplot(1, 5, 4)
sns.histplot(df_transf_diabetes['cuadratica'], kde=True)
plt.title('s5 cuadrática')

plt.subplot(1, 5, 5)
sns.histplot(df_transf_diabetes['boxcox'], kde=True)
plt.title(f'Box-Cox (lambda={lambda_boxcox:.2f})')

plt.tight_layout()
plt.show()
```

La transformación **log** reduce la cola derecha y estabiliza mejor la varianza, mientras la **log reflejada** invierte el sentido para explorar asimetría opuesta. La transformación **cuadrática** aumenta la separación en valores altos y puede amplificar la asimetría. **Box-Cox** busca automáticamente una potencia que aproxime normalidad, equilibrando forma y escala. Los gráficos para 's5' ilustran cómo logarítmica y Box-Cox intentan normalizar la distribución, mientras que la cuadrática la distorsiona. Box-Cox se muestra como la opción más efectiva para reducir la asimetría de 's5', preparando la variable para modelos sensibles a la distribución.

4. Aplica las siguientes **transformaciones** sobre una de las variables seleccionadas:

   - **Raíz cuadrada** usando `np.sqrt()`  
   - **Recíproca** usando `1/x`  
   - **Transformación Yeo-Johnson** usando `PowerTransformer(method="yeo-johnson")`

5. Genera **gráficas comparativas** que permitan visualizar la distribución de la variable en los siguientes escenarios:

   - Distribución con **transformación raíz cuadrada**

   - Distribución con **transformación Yeo-Johnson**

6. Escribe un análisis de **máximo cinco líneas** en el que expliques las diferencias entre las técnicas o métodos, su impacto en los resultados y cuál consideras más recomendable, justificando brevemente tu elección.

```python
from ucimlrepo import fetch_ucirepo

heart_disease = fetch_ucirepo(id=45)
df_heart = heart_disease.data.features.copy()
df_heart.head()
```

```python
df_heart.info()
```

## Variables escogidas

- chol
- trestbps
- thalach

```python
from sklearn.preprocessing import PowerTransformer

heart_vars = ['chol', 'trestbps', 'thalach']
df_heart_num = df_heart[heart_vars].apply(pd.to_numeric, errors='coerce').dropna().reset_index(drop=True)

x_heart = df_heart_num['chol'].copy()

x_raiz = np.sqrt(np.clip(x_heart, a_min=0, a_max=None))
x_reciproca = 1 / (x_heart + 1e-6)

pt = PowerTransformer(method='yeo-johnson', standardize=False)
x_yeojohnson = pt.fit_transform(x_heart.to_numpy().reshape(-1, 1)).flatten()

df_transf_heart = pd.DataFrame({
    'original': x_heart,
    'raiz_cuadrada': x_raiz,
    'reciproca': x_reciproca,
    'yeo_johnson': x_yeojohnson
})

df_transf_heart.head()
```

## Gráficos comparativos para chol

```python
plt.figure(figsize=(16, 5))

plt.subplot(1, 4, 1)
sns.histplot(df_transf_heart['original'], kde=True)
plt.title('Original chol')

plt.subplot(1, 4, 2)
sns.histplot(df_transf_heart['raiz_cuadrada'], kde=True)
plt.title('Raíz cuadrada')

plt.subplot(1, 4, 3)
sns.histplot(df_transf_heart['reciproca'], kde=True)
plt.title('Recíproca')

plt.subplot(1, 4, 4)
sns.histplot(df_transf_heart['yeo_johnson'], kde=True)
plt.title('Yeo-Johnson')

plt.tight_layout()
plt.show()
```

La transformación de **raíz cuadrada** reduce moderadamente la asimetría positiva en 'chol'. La **recíproca** comprime valores altos y puede exagerar diferencias en valores bajos, resultando en una distribución muy sesgada. **Yeo-Johnson** ajusta automáticamente la potencia, logrando una distribución más equilibrada y es robusta a ceros o valores negativos. Los histogramas de 'chol' confirman que la raíz cuadrada y Yeo-Johnson suavizan la cola derecha, mientras la recíproca introduce una fuerte asimetría izquierda. Yeo-Johnson suele ser la alternativa más estable y recomendable por su flexibilidad para lograr una distribución más simétrica.
