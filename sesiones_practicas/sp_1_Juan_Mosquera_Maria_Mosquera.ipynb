## **2. Descripción de la base de datos**

**Variables del dataset**



- Student_ID: Unique identifier for each student (S001–S500).
- Age: Student age (17–25), representing the Gen Z/Alpha transition.
- Gender: Male/Female demographic split.
- Daily_Screen_Time_Hours: Total hours spent on digital devices (The Independent Variable).
- Social_Media_Hours: Time spent on platforms like Instagram, LinkedIn, or X.
- Online_Study_Hours: Time spent on Coursera, YouTube Education, or live classes.
- Gaming_Hours: Time dedicated to mobile or PC gaming.
- Sleep_Hours: Average rest per night (A critical health metric).
- Attendance_Percentage: Percentage of classes attended (The "discipline" metric).
- Offline_Study_Hours: Traditional self-study with books and notes.
- Previous_Sem_CGPA: Baseline academic performance.
- Current_Sem_CGPA: The target variable for prediction and trend analysis.

# **4. Visualizaciones exploratorias básicas**

```python
sns.set(style="whitegrid")

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Screen Time
sns.histplot(
    df["daily_screen_time_hours"],
    kde=True,
    ax=axes[0, 0]
)
axes[0, 0].set_title("Distribución: Screen Time")

# Gaming
sns.histplot(
    df["gaming_hours"],
    kde=True,
    ax=axes[0, 1],
    color="red"
)
axes[0, 1].set_title("Distribución: Gaming Hours")

# Online Study
sns.histplot(
    df["online_study_hours"],
    kde=True,
    ax=axes[1, 0],
    color="green"
)
axes[1, 0].set_title("Distribución: Online Study Hours")

# Sleep
sns.histplot(
    df["sleep_hours"],
    kde=True,
    ax=axes[1, 1],
    color="purple"
)
axes[1, 1].set_title("Distribución: Sleep Hours")

plt.tight_layout()
plt.show()
```

```python
sns.boxplot(data=df[["daily_screen_time_hours",
                     "gaming_hours",
                     "online_study_hours",
                     "sleep_hours"]])
plt.show()
```

```python

sns.set(style="whitegrid")

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

#  Screen Time vs CGPA
sns.regplot(
    data=df,
    x="daily_screen_time_hours",
    y="current_sem_CGPA",
    scatter_kws={"alpha":0.5},
    ax=axes[0, 0]
)
axes[0, 0].set_title("Screen Time vs CGPA")

# Gaming vs CGPA
sns.regplot(
    data=df,
    x="gaming_hours",
    y="current_sem_CGPA",
    scatter_kws={"alpha":0.5},
    color="red",
    ax=axes[0, 1]
)
axes[0, 1].set_title("Gaming Hours vs CGPA")

# Online Study vs CGPA
sns.regplot(
    data=df,
    x="online_study_hours",
    y="current_sem_CGPA",
    scatter_kws={"alpha":0.5},
    color="green",
    ax=axes[1, 0]
)
axes[1, 0].set_title("Online Study vs CGPA")

# Sleep vs CGPA
sns.regplot(
    data=df,
    x="sleep_hours",
    y="current_sem_CGPA",
    scatter_kws={"alpha":0.5},
    color="purple",
    ax=axes[1, 1]
)
axes[1, 1].set_title("Sleep Hours vs CGPA")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,8))
corr = df.corr(numeric_only=True)

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Matriz de Correlación")
plt.show()
```

# 5. Interpretaciones iniciales

## 5.1 ¿Qué tendencias o patrones básicos se observan?

A partir de las visualizaciones obtenidas (gráficos de regresión, distribuciones y matriz de correlación), se identifican los siguientes patrones:

### Relación entre tiempo en pantalla y rendimiento académico
Se observa una relación negativa fuerte entre `daily_screen_time_hours` y `current_sem_CGPA`. A medida que aumentan las horas frente a la pantalla, el promedio académico disminuye de forma casi lineal.

### Gaming y rendimiento
Existe una tendencia negativa pronunciada entre `gaming_hours` y el CGPA actual. La pendiente descendente en el gráfico de regresión sugiere que un mayor tiempo dedicado a videojuegos se asocia con menor desempeño académico.

### Estudio online y rendimiento
La variable `online_study_hours` presenta una relación positiva fuerte con el rendimiento académico. A mayor tiempo dedicado al estudio digital, mayor es el CGPA observado.

### Sueño y rendimiento
Las horas de sueño (`sleep_hours`) muestran también una relación positiva clara con el desempeño académico. Los estudiantes que duermen más tienden a presentar mejores resultados.

### Estructura general de correlaciones
La matriz de correlación revela coeficientes extremadamente altos (|r| > 0.80) entre varias variables académicas y conductuales. Incluso se observan correlaciones cercanas a 1.00 entre variables como:
- `attendance_percentage`
- `offline_study_hours`
- `previous_sem_CGPA`
- `current_sem_CGPA`

Esto sugiere una relación lineal casi perfecta entre algunas variables del conjunto de datos.

---

## 5.2 ¿Qué limitaciones tiene la base de datos?

Del análisis visual y estadístico se identifican las siguientes limitaciones:

### a) Correlaciones excesivamente altas
En estudios de comportamiento humano, correlaciones superiores a 0.80 son poco frecuentes. La presencia de valores tan elevados sugiere:

- Posible generación sintética de datos.
- Dependencia matemática directa entre variables.
- Escasa presencia de ruido natural en las observaciones.

### b) Multicolinealidad extrema
Variables como `previous_sem_CGPA`, `attendance_percentage` y `offline_study_hours` presentan correlaciones casi perfectas con `current_sem_CGPA`, lo que implica:

- Redundancia informativa.
- Riesgo de sobreajuste en modelos predictivos.
- Dificultad para interpretar efectos individuales.

### c) Distribuciones poco naturales
Las distribuciones observadas muestran patrones relativamente estructurados y con dispersión limitada, lo que reduce la complejidad real del fenómeno estudiado.

### d) Baja variabilidad demográfica
La variable `Age` presenta correlaciones cercanas a cero con el resto de variables, lo que sugiere una limitada capacidad explicativa dentro del rango de edad analizado (17–25 años).

## 5.3 ¿Qué hipótesis iniciales podrían plantearse?

A partir del análisis exploratorio, se pueden formular las siguientes hipótesis preliminares:

**H1:** A mayor tiempo total frente a pantallas digitales, menor rendimiento académico.

**H2:** El tiempo dedicado al estudio online tiene un efecto positivo significativo sobre el desempeño académico.

**H3:** Las horas de sueño influyen positivamente en el rendimiento académico.

**H4:** Existe una fuerte relación entre disciplina académica (asistencia y estudio offline) y desempeño final.

## Conclusión preliminar

Si bien las tendencias observadas son coherentes desde un punto de vista lógico, la estructura casi perfectamente lineal del conjunto de datos limita su valor real. La alta colinealidad y la fuerza excesiva de las correlaciones reducen la capacidad de explorar relaciones complejas y realistas.

Por lo tanto, resulta pertinente considerar en una próxima sesión práctica el uso de un dataset con mayor variabilidad natural y menor dependencia estructural entre variables, que permita aplicar técnicas de análisis más robustas y obtener conclusiones con mayor validez externa.
